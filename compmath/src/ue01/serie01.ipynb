## Problem 1. Breaking down integers

```python
import math
import time
```

### 1a) Implementation

```python
def intbreak(n):
    return [k+1 for k in range(n)]
```

### 1a) Testing

```python
intbreak(10)
```

### 2a) Implementation

```python
start = time.time()
p_start = time.process_time()
def reorder(n):
    x = [math.floor(n/(10**i))%10 for i in range(math.ceil(math.log(n,10))-1,-1,-1)]
    magic = lambda nums : int(''.join(str(i) for i in sorted(x, key = lambda x : x % 2 == 0)))
    return magic(x)
p_end = time.process_time()
end = time.time()
print(f"Wall time = {end-start}".format())
print(f"Process time = {p_end - p_start}")
```

```python
def reorder2(n):
    # first we convert n into list by list comprehension
    int_as_list = [math.floor(n/(10**i))%10 for i in range(n)]
    return int_as_list
reorder2(364812)
```
